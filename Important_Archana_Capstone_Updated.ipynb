```python
# Split up the dataset
Rating_high = TV_df[(TV_df['Rating'] > 4)]
Rating_high.info()
print()
Rating_low = TV_df[(TV_df['Rating'] < 4)]
Rating_low.info()
```

```python
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(8, 4), sharey=True)

# Rating High distribution
ax1.hist(Rating_high['Rating'], alpha=0.5, label="Rating_high", color='blue')
ax1.set_title("Rating_high")
ax1.set_xlabel("Rating")
ax1.set_ylabel("Brand")


# rating_ Low distribution
ax2.hist(Rating_low['Rating'], alpha=0.5, label="Rating_low", color='orange')
ax2.set_title("Rating_low")
ax2.set_xlabel("Rating")


# Show the plot
plt.show()
```

```python
# Check for normality using skewness and kurtosis
print(stats.describe(Rating_high['Rating']))
print(stats.describe(Rating_low['Rating']))
```

## perform the Shapiro-Wilk test

```python
from scipy.stats import shapiro

# perform Shapiro-Wilk test
stat, p = shapiro(Rating_high['Rating'])

print('Statistics=%.3f, p=%.3f' % (stat, p))

# interpret
alpha = 0.05
if p > alpha:
    print('Data appears Gaussian (fail to reject H0)')
else:
    print('Data does not appear Gaussian (reject H0)')
```

The Shapiro-Wilk test provides a p-value of 0.000, which is less than the standard significance level (alpha) of 0.05. As a result, the null hypothesis is rejected and we can conclude that the data does not appear to be normally distributed.

## Now lets do a data transformation

```python
from scipy.stats.mstats import winsorize

# Apply Winsorization
Rating_high['Rating_winsorized'] = winsorize(Rating_high['Rating'], limits=[0.05, 0.05])

# Check the result of Winsorization with a boxplot or histogram
Rating_high['Rating_winsorized'].hist()

# Re-run Shapiro-Wilk test on Winsorized data
stat, p = shapiro(Rating_high['Rating_winsorized'])

print('Statistics=%.3f, p=%.3f' % (stat, p))

# interpret
alpha = 0.05
if p > alpha:
    print('Data appears Gaussian (fail to reject H0)')
else:
    print('Data does not appear Gaussian (reject H0)')
```

The results show that even after Winsorization, the distribution of 'Rating_high' does not appear to follow a Gaussian (normal) distribution, as the p-value is less than the significance level (alpha = 0.05), and thus we reject the null hypothesis of the Shapiro-Wilk test.

# When the data do not meet the assumption of normality, non-parametric tests can be a good choice as they do not require this assumption. Let's use the Mann-Whitney U test, which is a non-parametric equivalent to the independent samples t-test.


```python
from scipy.stats import mannwhitneyu

# perform Mann-Whitney U test
stat, p = mannwhitneyu(Rating_high['Rating'], Rating_low['Rating'])

print('Statistics=%.3f, p=%.3f' % (stat, p))

# interpret
alpha = 0.05
if p > alpha:
    print('Same distribution (fail to reject H0)')
else:
    print('Different distribution (reject H0)')

```

The p-value is essentially 0 (1.0253424610155138e-64 is extremely close to zero), which is less than the typical threshold of 0.05. This means that there is strong evidence to reject the null hypothesis (that the distributions of the two groups are the same). Therefore, you can conclude that there's a significant difference between the ratings of the high and low groups

TV brands are not rated similarly, i.e., the distributions of ratings for the high and low categories are significantly different. It implies that the ratings received by TV brands do indeed impact their distribution, and potentially their sales, customer preference, and other related factors.
